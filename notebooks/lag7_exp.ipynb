# Greenhouse Gas Concentration Analysis: Investigate bi-monthly autocorrelation structure  
**Data Source:** 
NOAA Global Monitoring Laboratory ([gml.noaa.gov](https://gml.noaa.gov/data/data.php?site=MLO&category=Greenhouse%2BGases))  

**Experiment:**  
I found a ~ bi-monthly autocorrelation signature in the CH4 project that is focused on flask data from the Mauna Loa Observatory (MLO) in Hawaii.  Literature searches have lead me to consider that this bi-monthly signature is partially or completely due to the Madden-Julian Oscillation (MJO).  

To cleanly isolate whether the 8-week signal is driven by tropical wave transport, high-elevation free-troposphere mechanics, or local Pacific dynamics, I will take the following approach:

Experimental Framework:

Target Site      
- MLO (Subtrop, High Alt)          

Positive Controls
- KUM (Local sea)
- SMO (Core tropical)
- GMI (Core tropical)

Negative Controls
- SPO (Antarctic ice)
- NWR (Mid-Lat high)

1. The Target Site
- MLO (Mauna Loa, Hawaii): (19.53 N, 3,397 m). Subtropical free troposphere.
  
2. Positive Controls (Should show the ~8-week signal)
- SMO (Cape Matatula, American Samoa): (14.25 S, 42 m). Keeps the model anchored in the tropical Pacific. This confirms if the 8-week signal is a broad, basin-wide tropical Pacific atmospheric feature.  This is a positive control for tropical intraseasonal transport (MJO, MISO).  
- GMI (Mariana Islands, Guam): (13.38 N, 6 m). Located directly downwind of the Asian monsoon outflow jet. If the MLO signal is heavily influenced by Monsoon Intraseasonal Oscillations (MISO) transport, TAP should show a highly correlated, slightly phased-advanced 6–8 week signature.
  
3. Negative Controls (Should NOT show the 8-week signal)
- SPO (South Pole, Antarctica): (90.00 S, 2,810 m). Keeps the model grounded with a pristine, polar, high-altitude baseline completely isolated from the tropics.  This is a negative control.  If the same bi-monthly dependence is observed, it would suggest the need to investigate the possibility that the signal is a processing artifact or that there is a broader atmospheric mechanism beyond tropical oscillations.  
- KUM (Cape Kumukahi, Hawaii): 19.52 N, 3 m). It matches MLO's latitude exactly and sits just a few miles away, but at sea level in the Marine Boundary Layer.
    - The Test: KUM is essentially the same longitude and latitude of MLO, but at a much lower altitude.  So, this is testing boundary layer versus free troposphere. If MLO shows the 8-week signal and KUM does not, this would strongly suggest that the signal relies strictly on high-altitude free-tropospheric transport or localized mountain meteorology.
- NWR (Niwot Ridge, Colorado): (40.05 N, 3,523 m). Crucial substitution for elevation matching. It matches MLO’s high elevation and free-tropospheric exposure but sits at a completely different, continental mid-latitude outside the tropical wave belt.
    - The Test: If MLO shows the signal and NWR does not, this would highly suggest that the 8-week signal is a tropical/subtropical phenomenon, rather than an inherent structural feature of high-altitude mountain datasets.

**Key Notes**  
- **`date` column**: datetime information in datetime64[ns] with dropped timezone info.
- **`value` column**: Gas concentration in dry air: ppb (nmol/mol)  
- **`value_unc` column**: The estimated uncertainty in reported value (in nanomol/mol (ppb)). Missing data coded as -999.999.  The corresponding value column value for missing data is -999.99.
- **`qcflag` column**: 3-character flags to indicate retained or rejected flask results as follows:

If the first character is not a period, the sample result should be
rejected for scientific use due to sample collection and/or measurement
issues. A second column character other than a period indicates a sample
that has no identifiable measurement or sampling errors but does not meet selection for representativeness
such as midday sampling or background air sampling or is otherwise atypical for a given sampling location and season. A third column flag
other than a period indicates noteworthy circumstances that are not known
to affect the data quality, but may have potential to.

| Indication | Flag | Description |
|------------|------|-------------|
| Retained |        ... |     good pair, no other issues |
| Retained |        ..* |     good pair, no definitive issues |
| Rejected |        M.. |     sample measurement issue |
| Rejected |        C.. |     sample collection issue |
| Rejected |        B.. |     both measurement and collection issues |
| Selection |       .S. |     selection issue. High/low mole fraction thought to not represent background conditions for example. |
| Informational |   ..M |     informational measurement tag or potential measurement issue |
| Informational |   ..C |     informational collection tag or potential collection issue |

- **`method` column**: single-character code used to identify the sample collection method as follows:

| Code | Description |
|------|-------------|
| P |Sample collected using a portable, battery powered pumping unit.  Two flasks are connected in series, flushed with air, and then pressurized to 1.2 - 1.5 times ambient pressure. |
| D | Similar to P but the air passes through a condenser cooled to about 5 deg C to partially dry the sample. |
| G | Similar to D but with a gold-plated condenser. |
| T | Evacuated flask filled by opening an O-ring sealed stopcock. |
| S | Flasks filled at NOAA GML observatories by sampling air from the in situ CO2 measurement air intake system. |
| N | Before 1981, flasks filled using a hand-held aspirator bulb. After 1981, flasks filled using a pump different from those used in method P, D, or G. |
| F | Five liter evacuated flasks filled by opening a ground glass, greased stopcock |

# Load datasets

```python
# define dataset URLs
datasets = {
    'MLO': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_mlo_surface-flask_1_ccgg_event.txt',
    'KUM': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_kum_surface-flask_1_ccgg_event.txt',
    'SMO': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_smo_surface-flask_1_ccgg_event.txt',
    'SPO': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_spo_surface-flask_1_ccgg_event.txt',
    'NWR': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_nwr_surface-flask_1_ccgg_event.txt',
    'GMI': 'https://gml.noaa.gov/aftp/data/trace_gases/ch4/flask/surface/txt/ch4_gmi_surface-flask_1_ccgg_event.txt'
}

# load datasets
df_list = []

for site, url in datasets.items():
    # comment = '#' automatically strips NOAA's metadata header lines
    # delim_whitespace = True handles the space-separated columns
    df = pd.read_csv(url, comment='#', sep=r'\s+', header=0)

    # track which observation site this row cam from
    df['site_code'] = site

    df_list.append(df)
    print(f"Loaded {site}: {df.shape[1]} columns, {df.shape[0]} rows.")

# verify header uniformity
# extract the baseline columns from the first DataFrame (MLO) in the list
first_columns = list(df_list[0].columns)
mismatched_sites = []

# compare every DataFrame's columns to the baseling (MLO)
for i, df in enumerate(df_list):
    if list(df.columns) != first_columns:
        # if they don't match, map the index back to the site code string
        site_code = list(datasets.keys())[i]
        mismatched_sites.append(site_code)

if not mismatched_sites:
    print("\nSuccess: All datasets have matching features.")
else:
    print(f"\nWarning: The following sites have different schemas: {mistmatched_sites}")
```

# Concatenate and clean

```python
# concatenate the df_list list
df_sites_raw = pd.concat(df_list, ignore_index=True)

# Keep features
cols_to_keep = ['site_code', 'datetime', 'value', 'value_unc', 'latitude', 'longitude', 'altitude', 'qcflag']
df = df_sites_raw[[c for c in cols_to_keep if c in df_sites_raw.columns]].copy()

# format/normalize the datetime values to YYYY:MM:DD
df['date'] = pd.to_datetime(df['datetime'], errors='coerce').dt.tz_localize(None).dt.normalize() # remove the time info
df.drop(columns=['datetime'], inplace=True)

n_bad_dates = df['date'].isna().sum()
if n_bad_dates > 0:
    print(f"Warning: {n_bad_dates} rows with unparseable datetimes - inspect before dropping.")

# filter the data for quality control problems:
# i.e. qcflag != '...', value_unc >= 3, value = -999.999 or -999.99
qc_mask = (
    (df['qcflag'] == '...') &
    (df['value_unc'] <= 3) &
    (~df['value'].isin([-999.999, -999.99]))
)

df_clean = df[qc_mask].copy()

print(f"Raw observations (all sites): {len(df)}")
print(f"filtered observations: {len(df) - len(df_clean)}")
print(f"Retained after QC: {len(df_clean)}")
print()
print(df_clean.groupby('site_code')['value'].count().rename('n_clean'))

# weekly aggregation
# handle the multiple and inconsistent number of sample collections per week
# assign each observation to its ISO week period before grouping
# grouping by both site_code and year_week keeps the six site series independent
df_clean['year_week'] = df_clean['date'].dt.to_period('W')

df_weekly = (
    df_clean
    .groupby(['site_code', 'year_week'], sort=True)
    .agg(
        ch4 = ('value', 'median'),
        # aggregate the errors (value_unc) using the propagation of uncertainty rule
        ch4_unc = ('value_unc', lambda x: np.sqrt((x**2).sum()) / len(x)),  
        n_flasks = ('value', 'count'),
        lat = ('latitude', 'first'),
        lon = ('longitude', 'first'),
        alt_masl = ('altitude', 'first'),
    )
    .reset_index()
)

# Convert ISO period to week-start timestamp (Monday) for time calculatins and plotting
df_weekly['date'] = df_weekly['year_week'].dt.start_time
df_weekly.drop(columns='year_week', inplace=True)
df_weekly.sort_values(['site_code', 'date'], inplace=True)
df_weekly.reset_index(drop=True, inplace=True)

print()
print(df_weekly.groupby('site_code')[['ch4']].agg(['count', 'median']).round(2))

# NOTE:  ch4_unc propagation uses the quadrature formula: sqrt(sum of squares) / n. 
# This gives the standard error of the median under the assumption that individual flask uncertainties are independent, 
# which is a reasonable approximation.
```

# Preprocessing loop

```python
# site-level preprocessing
# store the cleaned series and fitted preprocessors keyed by site code
# this allows access to both the output and fitted object 

preprocessors = {}  # site --> fitted GasPreprocessor instance
cleaned_series = {}  # site --> pd.Series (weekly, cleaned, interpolated)

site_params = {
    # use the same settings as the the MLO project
    'MLO': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
    'KUM': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
    'SMO': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
    'GMI': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
    'SPO': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
    'NMR': dict(seasonal_period=52, window=7, iqr_factor=1.5, interpolate_method='linear', lags=52, do_eda=True), 
}

for site, params in site_params.items():
    print(f"\n{'='*60}")
    print(f"   Preprocessing site: {site}")
    print(f"\n{'='*60}")

    # slice this site's weekly data and pass only the columns the processor needs
    site_df = (
        df_weekly[
            df_weekly['site_code'] == site]
            [['date', 'ch4']]
            .sort_values('date')
            .reset_index(drop=True)
    )

    if len(site_df) < 104:  # less than ~ 2 years, STL won't be reliable
        print(f"[WARNING] {site} has only {len(site_df)} weekly observations - skipping.")
        continue

    preprocessor = GasPreprocessor(gas_name='ch4', **params)

    try:
        series = preprocessor.fit_transform(
            site_df,
            custom_title=f"{site}: CH4 Raw vs. Cleaned"
        )
        preprocessors[site] = preprocessor
        cleaned_series[site] = series
        print(f"[OK] {site}: {len(series)} clean weekly observations, " 
              f"{series.index.min().date()} to {series.index.max().date()}")
    except Exception as e:
        print(f"[ERROR] {site} preprocessing failed: {e}")

print(f"\nPreprocessing complete. Sites read: {list(cleaned_series.keys())}")
```
