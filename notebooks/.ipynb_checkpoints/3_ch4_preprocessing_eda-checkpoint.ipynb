# Resample the full dataset

The full dataset will be resampled to weekly ('W') ahead of train/test split and subsequent preprocessing because:
1.  The mode of the observations after aggregation is 52 observations/year (see notebook 2_ch4_eda_clean) 
2.  Previous efforts to resample the train and test sets separtely, as part of the GasPreprocessor class functionality, has led to problems with the gap between the last train set value and the first test set value.  I have experienced gaps that are either too great (2 week gap) or are non-existant such that the last train datapoint date is the same as the first test set datapoint date, which is data leakage.

```python
df_float = df[['value', 'value_unc']].resample('W').mean()
df_obj = df[['qcflag']].resample('W').agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

df_resamp = pd.concat([df_float, df_obj], axis=1)
```

The preprocessed cleaned CH4 data series is non-stationary according to the ADF and KPSS tests, which confirms the clear, positive trend and linear regression analysis of the yearly averaged data (slope = 5.5 ppb/year, see 3_ch4_eda_clean notebook).


STL decomposition stabilized error variance based on the Breusch-Pagan and White tests, which found no heteroscedasticity in the STL decomposition residuals.  

The ACF and PACF plots of the STL residuals look nearly identical.
- The sharp cutoff after lag 1 indicates either a smoothing effect from the STL decomposition or a genuine short-term, non-seasonal persistence. The fact that the PACF looks identical to the ACF suggests that no intermediate lags were removed.
- The significant, negative spikes at Lag 52:  STL with period=52 removes a deterministic seasonal component, but real atmospheric seasonality isn't perfectly fixed — the amplitude, timing, and shape of the annual CH4 cycle vary somewhat from year to year. A negative spike at lag 52 in the residuals reflects that the STL seasonal component doesn't perfectly capture year-to-year variation in the annual cycle.  The negative spike is likely the "left-over".
- lag 104: significant in the PACF but not the ACF plot. So, PACF persists accross multiples of the seaosonal period while ACF decays.  This suggests a seasonal (1,0,0,52) order.  

Starting SARIMA orders: 
- SARIMA(1,1,1)(1,0,0,52) considering sharp cut-off at lag 1 and 52 in both ACF and PACF
- SARIMA(0,1,0)(1,0,0,52) considering that the spike at lag 1 on both ACF and PACF are borderline signficant.  

The next steps are to determine if the seasonal cycles are multiplicative and to examine the ACF and PACF plots of the first-differenced and seasonally-differenced data to estimate the non-seasonal (p,q) and seasonal (P,Q) SARIMA orders. The differencing analysis will be done in the next notebook (4_ch4_modeling).  

# Non-seasonal differencing test

```python
# estimate the need for and number of non-seasonal with pmdarima "ndiffs"
# Purpose: Determines the d parameter for an ARIMA(p, d, q) model by removing primary trends, 
# random walks, or shifts in the mean over time.

d = ndiffs(ch4_preprocessed, test='kpss')
print(f'Estimated non-seasonal differences: d = {d}')
```

```python
# Verify that d=1 is the correct degree of differencing (minimum variance principle)

# difference the series
nsdiff1 = ch4_preprocessed.diff().dropna()
nsdiff2 = nsdiff1.diff().dropna()

# determine the variance of each series: d=0, d=1, d=2
var_d0 = np.var(ch4_preprocessed)
var_d1 = np.var(nsdiff1)
var_d2 = np.var(nsdiff2)

print(f'variance at d=0: {var_d0:.3f}')
print(f'variance at d=1: {var_d1:.3f}')
print(f'variance at d=2: {var_d2:.3f}')
```

**Interpretation of the ndiffs (KPSS) test**

- The ndiffs test returns d = 1 for the need and degree of differencing.
- The variance results confirm that d = 1 is the correct degree of differencing based on the minimal variance principle.  

# Seasonal differencing test

- Purpose: Determines the D parameter for a seasonal \(SARIMA(p, d, q)(P, D, Q)_m\) model.
- Target: Removes repeating cyclical patterns or seasonal fluctuations across a known period m (e.g., m=12 for monthly data).
- Underlying Tests: It evaluates seasonal stability using specialized seasonality tests like OCSB (Osborn-Chui-Smith-Birchenhall) or CH (Canova-Hansen).

```python
# Run the OCSB (Osborn, Chui, Smith, and Birchenhall) test to statistically determine if seasonal differencing is required

# estimate seasonal differences using OCSB test automatically
D = nsdiffs(ch4_preprocessed, m=52, test='ocsb')

print(f'Seasonal differences needed: D = {D}')
```

```python
# The global trend is much more substantial than the amplitudes of the seasonal cycles.

# estimate seasonal differences using OCSB test where d=1
D = nsdiffs(nsdiff1, m=52, test='ocsb')

print(f'Seasonal differences needed (D): {D}')
```

**Interprtetation of nsdiffs (ocsb) test**

the ndiffs(OCSB) test on the preprocessed CH4 data and the first differenced series both recommend D = 0, meaning that the seasonality is largely deterministic.

There is a single datapoint for 2025 causing the outlier of amplitude = 0 seen in the plot above. Boundary artifacts from the resample('W') aggregation pushes the last Sunday of 2024 
to Jan 5, 2025 (due to the ISO week boundary not aligning with the calendar year boundary). So, I will need to apply a count filter to exclude this edge case.  

Another observation: 1991 shows an amplitude of 64.76 ppb and 1991 has exactly 52 observations, so it's a complete year — but the amplitude is unusually high compared to the cluster of other years.

```python
print(seasonal_component[seasonal_component.index.year == 1991].describe())
```

From Claude:
The 1991 amplitude of 64.76 ppb is simply max - min = 36.67 - (-28.08) = 64.75 ppb, which checks out arithmetically. The question is whether those extreme values are real or edge-effect artifacts.

The mean of 0.08 ppb is essentially zero, which is correct — STL seasonal components are zero-centered by construction. The std of 15.54 ppb and the range of ~65 ppb are notably larger than the typical year (most years cluster around 41–55 ppb in your sorted output), suggesting the STL boundary effect is inflating the amplitude estimate for 1991.

This is a known STL behavior — the smoother has no data to the left of the series start, so it extrapolates, which can produce exaggerated seasonal estimates in the first and last one to two years. The practical fix for the amplitude analysis is simply to exclude 1991 along with 2025:

```python
# plot amplitude of the seasonal cycle vs mean CH4 concentration.  
# if the seasonal cycles do not scale with the long term mean, then the cycles are additive
# if the seasonal cycles scale with the mean, then the cycles are multiplicative and would benefit 
# from transformation.  

# get the stl_results.  
# stl_result is stored as an instance attribute
stl_result = preprocessor.stl_result_

# Group raw data by year to compute the annual mean
trend_component = stl_result.trend
yearly_mean = trend_component.groupby(trend_component.index.year).mean()

# Extract yearly amplitudes from the STL decomposition result
seasonal_component = stl_result.seasonal
yearly_amplitudes = seasonal_component.groupby(seasonal_component.index.year).agg(lambda x: x.max() - x.min())

# Calculate the single average seasonal amplitude for your legend
avg_amp = yearly_amplitudes.mean()

# Align the data (ensures both series match the exact same years, drop missing values)
# exclude boundary years weith known edge effects or partial data (see comments above)
boundary_years = [1991, 2025]
obs_per_year = seasonal_component.groupby(seasonal_component.index.year).count()
complete_years = obs_per_year[
    (obs_per_year >= 50) & (~obs_per_year.index.isin(boundary_years))
    ].index

combined = pd.DataFrame({
    'mean'     : yearly_mean,
    'amplitude': yearly_amplitudes
}).loc[complete_years].dropna()

# Extract x and y as 1D arrays for the regression line
x = combined['mean'].values
y = combined['amplitude'].values

# Calculate the regression line
m, b = np.polyfit(x, y, 1)

# Calculate the correlation coefficient (r)
corr = combined['mean'].corr(combined['amplitude'])

# Scatter plot: mean vs amplitude
plt.figure(figsize=(6,4))
plt.scatter(x, y, alpha=0.6)
plt.plot(x, m*x + b, color='red', linestyle='--', 
         label=f'Slope = {m:.3f}, $r$ = {corr:.2f}\nAvg Amplitude = {avg_amp:.2f} ppb')

plt.xlabel('Mean CH4 (ppb)')
plt.ylabel('Seasonal Amplitude (ppb)')
plt.title('Seasonal Amplitude vs. Mean CH4 Level (1991–2024)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.show()
```

**Interpretation:**

- slope = 0.074 means that for every 1 ppb increase in the mean CH4 level, the seasonal amplitude increases by 0.074 ppb. Therefore the seasonal cycles are increasing in magnitude as the CH4 concentration increases.  Over the ~250 ppb rise in the data range, that's a predicted amplitude increase of about 18.5 ppb, which is modest relative to the average amplitude of 49.32 ppb.

- correlation coefficient (r) = 0.61 suggests a weak to moderate positive linear relationship between the mean CH4 level and seasonal amplitude.  The r^2 value of 0.61^2 = ~0.37 indicates that approximately 37% of the variation in the seasonal amplitude is explained by the change in the mean CH4 level. The other ~63% of the amplitude variation is driven by something else, e.g. interannual variability in atmospheric cirucluation, etc.

r = 0.61 indicates a moderate relationship, but the slope of 0.074 is still small in absolute terms. A strongly multiplicative process would typically show amplitude scaling roughly proportionally with level — for that you'd expect slope ≈ amplitude/mean ≈ 49/1800 ≈ 0.027 as a proportionality constant, but with r approaching 0.9+. This slope (0.074) is somewhat larger than that proportionality constant but the r value (0.61) is well below that threshold.

Second, Breusch-Pagan and White tests both showed no heteroscedasticity in the STL residuals. Heteroscedasticity is the practical consequence of multiplicative seasonality that a log transform actually fixes — if it's not present in the residuals, the transform isn't solving a real problem in the modeling context.

Third, nsdiffs returning D=0 means the seasonal structure doesn't require seasonal differencing, which is another indicator that the seasonal component is behaving additively enough for standard modeling assumptions to hold.

Conclusion: transformation isn't warranted, however, a mild Box-Cox transformation will be considered if neccessary.  

## CH4 Preprocessing Results Summary

### **Preprocessing**
The preprocessing pipeline successfully handled outliers and missing (NaN) values that arose due to resampling while remaining consistent with the original time series characteristics.  

### **Stationarity Analysis**
All processing stages (raw --> smoothed --> resampled) show consistent non-stationarity:
- **ADF Test**: Fails to reject H0 (p > 0.05)
  - H0 (Null Hypothesis): The series has a unit root (it is non-stationary).
  - H1 (Alternative Hypothesis): The series is stationary.
- **KPSS Test**: Rejects H0 (p < 0.05)
  - H0 (Null Hypothesis): The series is stationary (around a constant or a deterministic trend).
  - H1 (Alternative Hypothesis): The series has a unit root (it is non-stationary).

*Interpretation*: Strong trend dominance persists through preprocessing, which is expected and appropriate.  ndiffs(kpss) indicates d = 1 to make the series stationary.  

### **STL Decomposition Insights**
Analysis of the STL decomposition residuals indicates homoscedastic variance:
- **Breusch-Pagan test**: No heteroscedasticity detected (p = 0.9850)
- **White test**: No heteroscedasticity detected (p = 0.9872)
  
Constant variance in residuals is an optimal characteristic for SARIMA modeling.  

The seasonal amplitude vs mean CH4 concentration reveals mild multiplicative tendency (slope = 0.074, r = 0.61).  However, since the Breusch-Pagan and White tests both are negative for heteroscedasticity and the nsdiffs(ocsb) indicates D = 0, there is no strong indication for transformation.  A mild Box-Cox transformation may be considered depending on model evaluations.      
  
### **Autocorrelation Diagnostics**
- non-seasonal autocorrelation:
  - Sharp cutoff after lag 1 on both ACF and PACF plots of the STL decomposition residuals suggests a short term memory autocorrelation in the irregular (residual) component.  This is likely best modeled with ARMA(1,1) non-seasonal orders in SARIMA or an AR(1) irregular in a UC (unobserved components, state space) model.
- seasonal autocorrelation: 
  - significant spike at lag 52 and 104 of the PACF and only lag 52 of the ACF plots of the STL decomposition residuals indicates that the deterministicc STL seasonal component didn't fully capture the seasonal variation.  Some seasonal structure remains in the residuals.  The leftover signal can come from two different sources: deterministic or stochastic seasonality.  A SARIMAX or UC model with deterministic Fourier terms (e.g. K = 3) may handle this residual signal.  If so, then the seasonal lags observed in the STL decomposition residuals were due to a specification mismatch artifact.  If the seasonal lags persist, then the signal is likely due to stochastic seasonality.  If so, then a stochastic seasonal component would need to be added to a UC model.  The nsdiffs result (D = 0) is evidence against stochastic seasonality, but not definitive.  Model parameterization results will need to be considered.  

### **Key Implications for Modeling**
1. Required transformations:  
   - **Differencing**: First order differencing will be necessary to address persistent (upward) trend
   - **Seasonal adjustment**: Seasonal terms (likely period 52) should be incorporated to capture remaining seasonal pattern
2. Initial model specification:  
   - ARIMA(1,1,1)(0-1,0,0-1) as a starting range of orders  
   - Seasonal periodicity: 52 weeks  
3. Variance stability:
   - The homoscedastic residuals suggest that variance-stabilizing transformations may not be necessary
   - However, the positive slope and r value of the seasonal amplitude vs annual mean CH4 suggest that a mild Box-Cox transformation may be beneficial.

## resample and interpolate value_unc and recombine into a DataFrame

**NOTE:** The following saved data file (`ch4_preprocessed_df.csv`) is preserved as a record of the fully cleaned and preprocessed dataset for reference and reproducibility.  

This dataset **will not be used directly for modeling or forecasting**, since doing so would risk data leakage.  
In the next notebook (`4_ch4_modeling`), the full preprocessed data will first be split into training and test sets.  
The training set will be used to fit and transform the `GasPreprocessor` class, and the resulting fitted preprocessor will then be applied to transform the test set before model evaluation (e.g., SARIMA modeling).

```python
value_unc_resampled = (
    df['value_unc']
    .resample('W-SUN')
    .mean()
    .interpolate(method='time')
    .ffill()
    .bfill()
)

# recombine with ch4_preprocessed data
ch4_preprocessed_df = pd.DataFrame({
    'ch4_ppb': ch4_preprocessed,
    'value_unc': value_unc_resampled.reindex(ch4_preprocessed.index)
}).reset_index(names='date')

ch4_preprocessed_df
```
