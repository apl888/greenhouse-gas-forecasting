```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sympy import *
```

```python
x = int(input('Choose a value of x to determine the value of y for y = 3x**3 + 7:  '))

y = 3 * x**3 + 7
print(y)
```

```python
# a function to return the f(x) = 2x + 1

def f(x):
    return (2 * x) + 1

x_vals = [0, 1, 2, 3, 4]
xy_vals = []

for x in x_vals:
    y = f(x)
    xy_vals.append({'x':x, 'y':y})
    
xy_vals_df = pd.DataFrame(xy_vals)
xy_vals_df.set_index('x', inplace=True)
xy_vals_df
```

```python
x_vals = [-9,-4,-3.3,-2.5,-1,0,1,2,2.5,2.7,3.3,4.1,9]
xy_vals = []

for x in x_vals:
    y = (x ** 2) + 1
    xy_vals.append({'x':x, 'y':y}) 

df = pd.DataFrame(xy_vals)
plt.plot(df['x'], df['y'], marker='o')
```

```python
from mpl_toolkits.mplot3d import Axes3D

x_vals = [0,1,2,3,4,5,6]
y_vals = [0,2,4,6,8,9,11]
xyres_vals = []

for x in x_vals:
    for y in y_vals:
        res = 2 * x + 3 * y
        xyres_vals.append({'x':x, 'y':y, 'res':res})

df = pd.DataFrame(xyres_vals)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

# ax.scatter(df['x'], df['y'], df['res'], c=df['res'])
surf = ax.plot_trisurf(df['x'], df['y'], df['res'])
```
