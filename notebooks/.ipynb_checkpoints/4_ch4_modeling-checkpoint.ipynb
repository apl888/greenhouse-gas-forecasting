# Load the Data

```python
df_combined = pd.read_csv('../data/processed/all_ghg_aligned_nan.csv',
                          parse_dates=['date'], index_col='date')
```

# Train/Test Split

```python
test_size = 52 # one year of weekly data

CH4_train, CH4_test = df_combined[['CH4']].iloc[:-test_size], df_combined[['CH4']].iloc[-test_size:]

print(f'Train data: {CH4_train.shape[0]} data points, from {CH4_train.index.min()} to {CH4_train.index.max()}')
print(f'Test data: {CH4_test.shape[0]} data points, from {CH4_test.index.min()} to {CH4_test.index.max()}')
```

## Proprocessing and EDA of CH4 Train Set

```python
# fit and transform the training data to the GasPreprocessor class
# transform the test data using the fitted preprocessor

# reset index to make 'date' column
CH4_train_with_date = CH4_train.reset_index()
CH4_test_with_date = CH4_test.reset_index()

CH4_train_preprocessor = GasPreprocessor(gas_name='CH4', do_eda=True)
CH4_train_preprocessed = CH4_train_preprocessor.fit_transform(
    CH4_train_with_date, 
    custom_title = 'CH4 Train (EDA)'
)

# transform the test data using the fitted preprocessor
CH4_test_preprocessed = CH4_train_preprocessor.transform(CH4_test_with_date)
```

The ACF plot has significant spikes at lags 1-5 (not including the first lag (lag 0)) and mildly oscillating behavior.  This suggests that autocorrelation exists in the training set after preprocessing.  Also notable are the significance of lags around 52 indicating seasonal autocorrelation.  

The gradual tapering down of significan lags in the ACF plot and the sharp cut off after lag 1 in the PACF plot suggest an AR(p) model, possibly AR(1).  

The ADF and KPSS results show that the residuals, and seasonal component, are both stationary, which is good.  However, the trend is non-stationary as expected.  So, differencing of the trend will likely be necessary for ARMA type models.  ALthough the seasonal comonent is stationary, I will still consider seasonal differencing (D = 1) if I use a SARIMA model.  

Heteroscedasticity - both the Breusch-Pagan and White tests indicate heteroscedasticity in the CH4 train set residuals.  So, log or Box-Cox transformation will be necessary.  

```python
print('CH4_train shape:', CH4_train_with_date.shape)
print('CH4_test shape:', CH4_test_with_date.shape)

print('CH4_train_preprocessed shape:', CH4_train_preprocessed.shape)
print('CH4_test_preprocessed shape:', CH4_test_preprocessed.shape)
```

The difference in numbers of observations is due to resampling (performed by the Preprocessor class).  

# Test Transformation & Differencing on the Train Set

## Log transformation & Differencing

```python
# log-transform the preprocessed CH4 training and test data
CH4_train_logged = np.log(CH4_train_preprocessed)
CH4_test_logged = np.log(CH4_test_preprocessed)
```

```python
# Test results of log transformation and differening of the training data

# check for heteroscesdasticity & stationarity in the log-transformed data
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_logged, label='Heteroscedasticity Test for Log-Transformed CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_logged, label='Log-Transformed CH4 Train Data')

# mean stabilization: differencing the logged data
CH4_train_logged_diff = CH4_train_logged.diff().dropna()

# run stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_logged_diff, label='Heteroscedasticity Test for Log-Transformed & Differenced CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_logged_diff, label='Log-Transformed & Differenced CH4 Train Data')

# seasonal differencing of the logged data
CH4_train_logged_s52diff = CH4_train_logged.diff(52).dropna()

# run heteroscedasticity & stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_logged_s52diff, label='Heteroscedasticity Test for Log-Transformed & Seasonally Differenced CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_logged_s52diff, label='Log-Transformed & Seasonally Differenced CH4 Train Data')

# non-seasonal and seasonal differencing of the logged data
CH4_train_logged_nsdiffs52diff = CH4_train_logged_diff.diff(52).dropna()

# run heteroscedasticity & stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_logged_nsdiffs52diff, label='Heteroscedasticity Test for Log-T & Non-Seasonally + Seasonally Diff CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_logged_nsdiffs52diff, label='Log-T & Non-Seasonally + Seasonally Diff CH4 Train Data')
```

Log transformation did not stabilize the variance in the CH4 training data.  This may not be an issue since the data range is small vs the absolute magnitude of the values.  The transformation affect needs to be assessed on the model residuals. It's possible that I will need to consider exogenous veriables (SARIMAX), SARIMA with generalized least squares (GLS), or modeling the variance with ARCH/GARCH. 

Importantly, taking the first difference (d=1) and seasonal difference (D=1) did make the data set stationary, so I don't need to test further degrees of differencing.  This will inform the SARIMA model parameters.

From the first non-seasonal difference and seasonal difference (lag = 52), 

- ADF and KPSS indicate that the series is stationary.
- ACF plot: slightly significant spike at lag 1 --> MA(q) = 1 or 2; significant spike at lag 52 suggests a seasonal MA(Q) = 1 (s = 52).
- PACF plot: very similar to the ACF plot with a slightly significant spike at lag 1 --> AR(p) = 1 or 2; also has a significant spike at lag 52, therefore AR(P) = 1.
- summary of SARIMA parameters to start with: (1/2,1,1/2) x (1,1,1)[52]

I will try pmdarima to auto-identify the ARIMA or SARIMA model parameters.

### Box-Cox Transformation & Differencing

```python
# I will try Box-Cox transformation.  
# First, make sure that the data is positive 

print((CH4_train_preprocessed <= 0).sum())
```

```python
# log-transform the preprocessed CH4 training data
CH4_train_boxcox, lambda_bc = boxcox(CH4_train_preprocessed)
```

```python
# Test results of Box-Cox transformation and differening of the training data

# wrap the CH4_train_boxcox in a pandas series since boxcox() returns a numpy array and my 
# _run_stationarity_and_ac_analysis() method expects a pandas series
CH4_train_boxcox = pd.Series(CH4_train_boxcox, index=CH4_train_preprocessed.index)

# check for heteroscesdasticity in the log-transformed data
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_boxcox, label='Heteroscedasticity Test for Box-Cox-Transformed CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_boxcox, label='Box-Cox-Transformed CH4 Train Data')

# mean stabilization: differencing the logged data
CH4_train_boxcox_diff = CH4_train_boxcox.diff().dropna()

# run stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_boxcox_diff, label='Heteroscedasticity Test for Box-Cox-Transformed & Differenced CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_boxcox_diff, label='Box-Cox-Transformed & Differenced CH4 Train Data')

# seasonal differencing of the logged data
CH4_train_boxcox_s52diff = CH4_train_boxcox.diff(52).dropna()

# run heteroscedasticity & stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_boxcox_s52diff, label='Heteroscedasticity Test for Box-Cox-Transformed & Seasonally Differenced CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_boxcox_s52diff, label='Box-Cox-Transformed & Seasonally Differenced CH4 Train Data')

# non-seasonal and seasonal differencing of the logged data
CH4_train_boxcox_nsdiffs52diff = CH4_train_boxcox_diff.diff(52).dropna()

# run heteroscedasticity & stationarity tests on the differenced data 
CH4_train_preprocessor.test_heteroscedasticity(CH4_train_boxcox_nsdiffs52diff, label='Heteroscedasticity Test for Box-Cox-T & Non-Seasonally + Seasonally Diff CH4 Train Data')
CH4_train_preprocessor._run_stationarity_and_ac_analysis(CH4_train_boxcox_nsdiffs52diff, label='Box-Cox-T & Non-Seasonally + Seasonally Diff CH4 Train Data')
```

```python
# print the lambda value for Box-Cox transformation
print(f'Box-Cox lambda value: {lambda_bc:.4f}')
```

The Box-Cox transformation yielded very similar results compared to the log transformed data.  So, I will stick with log transformation.

## Rolling Forecast

```python
print(CH4_train_logged.index.freq)
print(CH4_test_logged.index.freq)
print(CH4_train_preprocessed.index.freq)
print(CH4_test_preprocessed.index.freq)
```

```python
# rolling forecast of model 2: ARIMA(2,1,1)(2,0,1)[52]

# Initialize variables
CH4_train_logged.index.freq = 'W-SUN'
CH4_test_logged.index.freq = 'W-SUN'
CH4_train_preprocessed.index.freq = 'W-SUN'
CH4_test_preprocessed.index.freq = 'W-SUN'

history = CH4_train_logged.copy()  # Your initial training data
rolling_forecasts = []
rolling_conf_ints = []

# Rolling forecast loop
for t in range(len(CH4_test_preprocessed)):
    # Fit the model on the current history
    model = sm.tsa.SARIMAX(history, order=(2,1,1), seasonal_order=(2,0,1,52), trend='n')
    model_fit = model.fit(method='lbfgs', disp=False)
    
    # Forecast one step ahead
    forecast_result = model_fit.get_forecast(steps=1)
    
    # get the correct timestamp for the forecast
    forecast_index = CH4_test_preprocessed.index[t]
    
    # store the forecast and confidence intervals with explicit index
    forecast_mean = np.exp(forecast_result.predicted_mean.values[0])
    conf_int = np.exp(forecast_result.conf_int().values[0])
    
    # store the forecast and confident intervals with explicit index
    rolling_forecasts.append(pd.Series(forecast_mean, index=[forecast_index]))
    rolling_conf_ints.append(pd.DataFrame([[conf_int[0], conf_int[1]]], columns=['lower', 'upper'], index=[forecast_index]))
    
    # Update history with the actual observed test value (log-transformed)
    new_observation = CH4_test_logged.iloc[t]  # Still logged
    history = pd.concat([
        history, 
        pd.Series([new_observation], index=[CH4_test_preprocessed.index[t]])
        ])
    history = history.asfreq('W-SUN')  # Ensure the frequency is consistent 

# Convert to pandas Series/DataFrame
CH4_rolling_forecast2 = pd.concat(rolling_forecasts)
CH4_rolling_conf_int2 = pd.concat(rolling_conf_ints)
```

```python
# plot the rolling forecast results 

plt.figure(figsize=(10,6))
plt.plot(CH4_test_preprocessed.index, CH4_test_preprocessed, label='Actual (Test Set)', color='darkgreen') 
plt.plot(CH4_rolling_forecast2.index, CH4_rolling_forecast2, label='Rolling Forecast', color='blue') 
plt.fill_between(
    CH4_rolling_conf_int2.index, 
    CH4_rolling_conf_int2['lower'], 
    CH4_rolling_conf_int2['upper'], 
    color='lightblue', alpha=0.5, label='95% CI'
)
plt.legend()
plt.title('CH4 Concentration Rolling Forecast vs Actual: ARIMA(2,1,1)(2,0,1)[52]') 
plt.xlabel('Date')
plt.ylabel('[CH4] ppb')
plt.show()
```

```python
# check for missing or duplicate dates in each series

def check_timestamps(series, name):
    print(f'\nChecking {name}:')
    # check for duplicates
    duplicates = series.index.duplicated() 
    print(f'Duplicate timestamps: {duplicates.sum()}')
    
    # check for gaps 
    expected = pd.date_range(start=series.index[0], end=series.index[-1], freq='W-Sun')
    missing = expected.difference(series.index)
    print(f'Missing timestamps: {len(missing)}')
    if len(missing) > 0:
        print(f'Exmple missing dates : {missing[:5]}') 
        
check_timestamps(CH4_train_logged, 'CH4 Train Logged')
check_timestamps(CH4_test_logged, 'CH4 Test Logged')
check_timestamps(CH4_train_preprocessed, 'CH4 Train Preprocessed')
check_timestamps(CH4_test_preprocessed, 'CH4 Test Preprocessed')
```

## One-Step Ahead (Standard) Forecasting

```python
# Instead of forecasting the entire length of the test set, I will implement a one-step ahead, 
# standard SARIMA forecast and compare the results with the standard forecast (n = len(test set)), 
# and the rolling forecast (n = 1).  

# ensure that the frequency is set  
CH4_train_logged.index.freq = 'W-SUN'
CH4_test_logged.index.freq = 'W-SUN' 

# fit the model once, outside the loop (compared to rolling forecast)
model = sm.tsa.SARIMAX(
    CH4_train_logged, 
    order=(1,1,2), 
    seasonal_order=(1,1,1,52),
    enforce_stationarity=False,
    enforce_invertibility=False,
    trend='n' 
)
model_fit = model.fit(method='lbfgs', disp=False) 

# initialize the storage for the forecast and confidence intervals
one_step_forecasts = []
one_step_conf_ints = []

# one-step ahead forecast loop
for t in range(len(CH4_test_logged)):
    # get prediction for the next step (dynamic=False ensures one-step) 
    forecast_result = model_fit.get_prediction(
        start=CH4_test_logged.index[t], 
        end=CH4_test_logged.index[t], 
        dynamic=False           
    )
    
    # extract forecast and confidence intervals (log scale values) 
    forecast_mean_log = forecast_result.predicted_mean.iloc[0]
    conf_int_log = forecast_result.conf_int().values[0] # [lower, upper]
    
    # back-transform the forecast and confidence intervals to original scale
    forecast_mean = np.exp(forecast_mean_log)
    conf_int = np.exp(conf_int_log)
    
    # store the forecast and confidence intervals with explicit index
    one_step_forecasts.append(
        pd.Series(forecast_mean, index=[CH4_test_logged.index[t]])
    )
    one_step_conf_ints.append(
        pd.DataFrame([[conf_int[0], conf_int[1]]], 
                     columns=['lower', 'upper'], 
                     index=[CH4_test_logged.index[t]])
    )
    
# combine results
CH4_one_step_forecast8 = pd.concat(one_step_forecasts)
CH4_one_step_conf_int8 = pd.concat(one_step_conf_ints)
```

### Autocorrelation analysis of rolling forecast model residuals

```python
# Calculate the residuals from the rolling forecast models 

# model 2: ARIMA(2,1,1)(2,0,1)[52] 
CH4_rolling_forecast2_residuals = (np.exp(CH4_test_logged) - CH4_rolling_forecast2) 
CH4_rolling_forecast2_residuals = pd.Series(
    CH4_rolling_forecast2_residuals, 
    index=CH4_test_preprocessed.index
)

# model 5: ARIMA(2,1,0)(1,0,2)[52]
CH4_rolling_forecast5_residuals = (np.exp(CH4_test_logged) - CH4_rolling_forecast5)
CH4_rolling_forecast5_residuals = pd.Series(
    CH4_rolling_forecast5_residuals, 
    index=CH4_test_preprocessed.index
)

# model 8: ARIMA(1,1,2)(1,1,1)[52]
CH4_rolling_forecast8_residuals = (np.exp(CH4_test_logged) - CH4_rolling_forecast8)
CH4_rolling_forecast8_residuals = pd.Series(
    CH4_rolling_forecast8_residuals, 
    index=CH4_test_preprocessed.index
)

```

PACF function uses the Yule-Walker method to calculate the partial autocorrelation function (OLS method).  To avoid overfitting and ensure numerical stability, statsmodels enforces that # lags < 50% of the # observations. Since my test set is 66 data points in length, I am limited to 32 lags.  Two possible ways forward with the residuals test for autocorrelation:

1. move forward as is and interpret ACF based on 52 lags and PACF based on 32 lags.  
2. Go back and change the test set to include 106 data points (nlags < len(residuals) // 2).  

I will choose option 1 and determine if I need to change the data split ratio after I see the results.

```python
fig, axs = plt.subplots(4,2, figsize=(12, 14)) 

plot_acf(CH4_rolling_forecast2_residuals, lags=52, ax=axs[0,0])
axs[0,0].set_title('ACF of Residuals: Rolling Forecast Model 2 (2,1,1)(2,0,1)[52]')

plot_pacf(CH4_rolling_forecast2_residuals, lags=32, ax=axs[0,1])
axs[0,1].set_title('PACF of Residuals: Rolling Forecast Model 2 (2,1,1)(2,0,1)[52]')

plot_acf(CH4_rolling_forecast5_residuals, lags=52, ax=axs[1,0])
axs[1,0].set_title('ACF of Residuals: Rolling Forecast Model 5 (2,1,0)(1,0,2)[52]')

plot_pacf(CH4_rolling_forecast5_residuals, lags=32, ax=axs[1,1])
axs[1,1].set_title('PACF of Residuals: Rolling Forecast Model 5 (2,1,0)(1,0,2)[52]')

plot_acf(CH4_rolling_forecast8_residuals, lags=52, ax=axs[3,0])
axs[3,0].set_title('ACF of Residuals: Rolling Forecast Model 8 (1,1,2)(1,1,1)[52]')

plot_pacf(CH4_rolling_forecast8_residuals, lags=32, ax=axs[3,1])
axs[3,1].set_title('PACF of Residuals: Rolling Forecast Model 8 (1,1,2)(1,1,1)[52]')

plt.tight_layout()
plt.show()
```

Models 2, 5, and 6 have a oscillatory behavior through ~ lag 15.  All lags are within the CI (95%) so aren't significant, but the pattern suggests that there is possible room to improve the model tune.  

```python
# Ljung-Box test

models = {
    'Model 2 (2,1,1)(2,0,1)[52]': CH4_rolling_forecast2_residuals,
    'Model 5 (2,1,0)(1,0,2)[52]': CH4_rolling_forecast5_residuals,
    'Model 8 (1,1,2)(1,1,1)[52]': CH4_rolling_forecast8_residuals   
}

for model_name, residuals in models.items():
    lb_test = acorr_ljungbox(residuals, lags=[7,15,26,52], return_df=True)
    print(f'\nLjung-Box Test Results for {model_name}:')
    print(lb_test)
    print('\n' + '-'*50)
```

There are no significant lags and the Ljung-Box p-values at various lags are all > 0.05 meaning that there is no autocorrelation remaining in the residuals and the models have captured the non-seasonal and seasonal structure in the data.  Therefore, I can choose the best model based on the forecast metrics.  

overall, model 5 has the best performance based on all metrics and analyses. It has the lowest AIC and BIC scores, has slightly lower MAE and MAPE scores, it is tied for the best R^2 value, it failed to reject the autocorrelation and heteroscedasticity null hypothises, and the ACF/PACF plots of residuals look clean.  

it's tempting to go with model 6 due to its simplicity, but the model failed to reject 
the Ljung-Box test null hypotheses for autocorrelation at lag 52.
